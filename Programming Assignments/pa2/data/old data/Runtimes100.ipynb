```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data100.csv")
dims = data.groupby(["dim"])
```

```python
plt.style.use("seaborn")

for dim, group in dims:
    ax = group.plot(x="cross", y="time", kind="bar", title=f"{dim}x{dim} Runtimes", logy=True, legend=False, figsize=(20,10), fontsize=16, rot=0)
    ax.title.set_size(24)
    ax.set_xlabel("Cross-over Point", fontsize=16)
    ax.set_ylabel("Time (ms)", fontsize=16)
```
